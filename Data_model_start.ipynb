```python
#Initial Setup
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('dass.csv')
df.head()
```

```python
df['country']=pd.factorize(df['country'], sort=True)[0]+1
df['age_group']=pd.factorize(df['age_group'], sort=True)[0]+1
df['major']=pd.factorize(df['major'], sort=True)[0]+1
df=df.drop(['dep_cond','anx_cond','str_cond'],axis=1)
df.head()
```

```python
# Analyze Depression only
X = df.drop(['Depression','Depression_cat','Anxiety','Anxiety_cat','Stress','Stress_cat'],axis=1)
y = df['Stress_cat']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

#Standardization-transforming values so mean is 0 and SD is 1
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)

knn=KNeighborsClassifier(n_neighbors=13)
knn.fit(X_train, y_train)
y_predict = knn.predict(X_test)
print(knn.score(X_test,y_test))
```
